#### Part 3: Fit a linear model

Use the `lm` command to fit a linear regression to the `D_RT` variable using the four predictors. Assign the output to a new variable called `ml1`. Use the `summary(ml1)` command to display the output of the model:


```r
ml1 <- lm(D_RT ~ Length + SUBTLWF + Concreteness_Rating + Emotional_Valence, full_data)

ml1
summary(ml1)

## insert code ehre
```

What is the intercept value and the slope for the word length predictor? Explain what these values means:
> The intercept value is 667.07 and the slope for word length is 26.42. This intercept means that if all predictors are 0, we expect the mean RT to be 667.07 (not sure because it's not specified, but I assume milliseconds?). The slope means that with each increase of wordlength of one unit (here, I assume a unit is one letter), the RT increases by 26.42.

What is the R-squared value? What does this mean?
> The R-squared value is a measure for how well the model is fitting the actual data. In our output, we see that it is only 0.03, which means that about 3% of the variance found in the response variable RT can be explained by the predictor variables. 

As you noticed, the R-squared value is really low. This is because there is a lot of individual differences in how fast people react. These differences cannot be explained by word characteristics. One way to account for individual differences is to summarize the data over participants, and obtain the average RTs for each word. Let's do this and refit the model.

First, use the `group_by` and `summarise` commands to calculate the average RT for each word. In order to keep the word characteristics, include those columns in the `group_by` call. E.g.: `group_by(D_word, Length, SUBTLWF, Concreteness_Rating, Emotional_Valence)`. Save the result to a new variable. Use the `heads` command to look at the data:

```r
summary_RT <- full_data %>% 
             group_by(D_word, Length, SUBTLWF, Concreteness_Rating, Emotional_Valence) %>% 
             summarise(
               mean_RT=mean(D_RT)) # Get a table of mean RT per word

head(summary_RT)
```

Now, fit the linear model again using the summarised data. Do the estimates of intercept and slope differ much from the previous model? How about the R-squared value? If so, why do you think that is the case?

```r
ml2 <- lm(mean_RT ~ Length + SUBTLWF + Concreteness_Rating + Emotional_Valence, summary_RT) 
summary(ml2)
```

> The intercept and slopes have only changed very slightly, but the R-squared value is higher now and indicates that instead of about 3%, now about 11% of the variance found in the response variable RT can be explained by the predictor variables. I think the intercept and slope have stayed almost the same because nothing has changed in these mean values, but the R squared has gotten a bit higher because we just aggregated the RTs over all participants and so the variance between participants no longer needs to be accounted for.

Finally, change the predictor for the SUBTLWF to be the log(SUBTLWF) value. Does the R-squared increase? Why/why not?

```r
ml3 <- lm(mean_RT ~ Length + log(SUBTLWF) + Concreteness_Rating + Emotional_Valence, summary_RT) 
summary(ml3)
```

> The R squared increased a bit from .11 to .16 after using the log transformated variable. I think this is because the original variable SUBTLWF had a very skewed distribution (see plots below) and the log transformation helped making it more applicable (and fit better) for the linear regression I ran.

```r
ggplot(full_data, aes(x=SUBTLWF)) + geom_density()

ggplot(full_data, aes(log(SUBTLWF))) + geom_density()
```
