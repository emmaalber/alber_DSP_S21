#### Conceptual Exercise 3
##### (a) Explain how k-fold cross-validation is implemented.
A dataset is randomly split into k different subsets (aka folds) of equal or approximately equal size. One fold is then used as the testing set (validation set) while the method is trained (fit) on the remaining folds. Next, the MSE of the attained model is computed for the held-out fold. This is then repeated with every fold being the validation set once, which results in k cycles of training and testing. In the end, an average of MSEs over all cycles is computed.
The number k is chosen by the statistician. If k = n, they're basically performing LOOCV, but usually k is smaller than n.  

##### (b) What are the advantages and disadvantages of k-fold crossvalidation relative to:
i. The validation set approach?

Apart from the computational costs being lower, I don't see any advantage of k-fold crossvalidation relative to the validation set approach after reading the book. A disadvantage of the validation set approach compared to k-fold crossvalidation is that yields worse (more variable) estimates for the test error rate, because the validation set approach only makes use of half the dataset available and tends to overestimate the test error rate.

ii. LOOCV?

If k < n, then an advantage of k-fold crossvalidation is that the method poses less computational problems and takes less time, especially if n is very large. Additionally, the variance in k-fold crossvalidation is lower compared to LOOCV. This is conceptually connected to a disadvantage of k-fold crossvalidation: Compared to LOOCV, the bias is higher compared to LOOCV. We see that this is a bias-variance trade off and k must be chosen in order to find the best compromise here.

(d) Write a for loop from i = 1 to i = n, where n is the number of observations in the data set, that performs each of the following steps:

i. Fit a logistic regression model using all but the ith observation to predict Direction using Lag1 and Lag2.

ii. Compute the posterior probability of the market moving up for the ith observation.

iii. Use the posterior probability for the ith observation in order to predict whether or not the market moves up.

iv. Determine whether or not an error was made in predicting the direction for the ith observation. If an error was made, then indicate this as a 1, and otherwise indicate it as a 0.

```r
error <- rep(0, dim(Weekly)[1])
for (i in 1:dim(Weekly)[1]) 
{fit.glm <- glm(Direction ~ Lag1 + Lag2, data = Weekly[-i, ],  family = "binomial")
    pred.up <- predict.glm(fit.glm, Weekly[i, ], type = "response") > 0.5
    true.up <- Weekly[i, ]$Direction == "Up"
    if (pred.up != true.up)
        error[i] <- 1}

error
sum(error)
```

(e) Take the average of the n numbers obtained in (d) in order to obtain the LOOCV estimate for the test error. Comment on the results.

```r
mean(error)
```

By using LOOCV, we get a test error rate of approx. 45%. This means that our model classified about 55% of the test data (which it hadn't been trained on) correctly.
